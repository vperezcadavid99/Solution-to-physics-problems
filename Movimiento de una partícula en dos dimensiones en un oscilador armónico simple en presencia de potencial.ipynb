# Movimiento de una partícula en dos dimensiones (x,y) en un oscilador armónico simple no isotrópico en un potencial

En la presente tarea se va a descibir el movimiento de una partícula en dos dimensiones (x,y) en un oscilador armónico simple no isotrópico en un potencial de la forma

$$V(x,y) = \frac{1}{2}mw^{2}_xx^{2}+\frac{1}{2}mw^{2}_yy^{2}\, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \ (1)$$

donde consideramos las frecuecnias $w_x$ y $w_y$ distintas.

Primero se encuentran las variables de acción $I_x$ y $I_y$ junto con la expresión de la energía en términos de estas.

Para este sistema el hamiltoniano, según la Ec (1) corresponde a

$$H =  \frac{1}{2m}\left [ p^{2}_x + p^{2}_y \right ] + \frac{1}{2}mw^{2}_xx^{2} + \frac{1}{2}mw^{2}_yy^{2}\, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \ (2)$$

En este orden de ideas, a partir de la Ec (2) se encuentra que la ecuación de Hamilton Jacobi independiente del tiempo corresponde a

$$\frac{1}{2}\left [ \left ( \frac{\partial W}{\partial x} \right )^{2}+\left ( \frac{\partial W}{\partial y} \right )^{2} \right ] + \frac{1}{2}mw^{2}_xx^{2} + \frac{1}{2}mw^{2}_yy^{2} = E \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \ (3)$$

Se propone ahora una solución de la forma

$$W = X(x) + Y(y) \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \ (4) $$

Luego, la Ec(3) se escribe como

$$ \left [ \frac{1}{2m}\left ( \frac{dX}{dx} \right )^{2} +\frac{1}{2}mw^{2}_xx^{2} \right ] + \left [ \frac{1}{2m}\left ( \frac{dY}{dy} \right )^{2} +\frac{1}{2}mw^{2}_yy^{2} \right ] = E \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (5)$$

Dado que el primer término de la Ec (4) depende sólo de $x$ y el segundo sólo de $y$ y su suma es una constante, cada uno de estos téminos deben ser entonces igualados a constantes; sean estas $\alpha_x$ y $\alpha_y$ respectivamente, de este modo se tiene que

$$ \left [ \frac{1}{2m}\left ( \frac{dX}{dx} \right )^{2} +\frac{1}{2}mw^{2}_xx^{2} \right ] = \alpha_x \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (6)$$

$$\left [ \frac{1}{2m}\left ( \frac{dY}{dy} \right )^{2} +\frac{1}{2}mw^{2}_yy^{2} \right ] = \alpha_y \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (7)$$

de este modo, se cumple que

$$ \alpha_x + \alpha_y = E$$

En este orden de ideas, integrando las Ec (6) y Ec (7) para $x$ y $y$ respectivamente se sigue:

$$ X = \int\sqrt{2m\alpha_x - m^{2}w^{2}_xx^{2}} dx \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (8) $$

$$ Y = \int\sqrt{2m\alpha_y - m^{2}w^{2}_yy^{2}} dy \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (9) $$

Luego, sustituyendo Ec (8) y Ec (9) en Ec (4) se consigle una integral completa de la forma

$$W = \int\sqrt{2m\alpha_x - m^{2}w^{2}_xx^{2}} dx + \sqrt{2m\alpha_y - m^{2}w^{2}_yy^{2}} dy \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (10)$$

Recordando ahora de las transformaciones canónicas que la derivada de la Ec (10) respecto a las coordenadas son los momentos, se consigue

$$p_x = \frac{∂W}{\partial x} = \sqrt{2m\alpha_x - m^{2}w^{2}_xx^{2}}$$
$$p_y = \frac{∂W}{\partial y} = \sqrt{2m\alpha_y - m^{2}w^{2}_yy^{2}}$$

A partir de estas dos últimas expresiones, se encuentras que las variables de accioción $I_x$ y $I_y$ son 

$$ I_x = \frac{1}{2\pi }\oint \sqrt{2m\alpha_x - m^{2}w^{2}_xx^{2}} dx $$
$$ I_y = \frac{1}{2\pi }\oint \sqrt{2m\alpha_y - m^{2}w^{2}_yy^{2}} dy $$

de aquí se sigue que

$$I_x = \frac{\alpha_x}{\omega_x}$$
$$I_y = \frac{\alpha_y}{\omega_y}$$

luego, las constantes inicialmente introducidas corresponden a

$$\alpha_x = \omega_x I_x \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (11) $$
$$\alpha_y = \omega_y I_y \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (12)$$

Sustituyendo las Ec (11) y Ec (12) en la expresión para la enegía total se consigue su expresión final

$$E =  \omega_x I_x + \omega_y I_y $$

Así mismo, la integral completa con las expresiones (11) y (12) es

$$W = \int\sqrt{2m\omega_x I_x - m^{2}w^{2}_xx^{2}} dx + \sqrt{2m \omega_y I_y - m^{2}w^{2}_yy^{2}} dy \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \ (13) $$




Acorde a la teoría  de la sección 49.9, vimos que las variables de ángulo pueden obtenerse a partir de la acción reducida como: 

$$ w = \frac{\partial S(q, I)}{\partial I} $$

Usando la ecuación (13) podemos encontrar la variables de ángulo $\phi_x$ y $\phi_y$: 


$$ \phi_x =  \frac{\partial S(q, I)}{\partial I_x}  = \int \frac{2m \omega_x dx}{\sqrt{2m \omega  I_x - m^2 \omega_x^2 x^2}} =  \int \frac{dx}{(2I_x/mw_x) - x{^2}} $$

Y similarmente:

$$ \phi_y =  \frac{\partial S(q, I)}{\partial I_y} =   \int \frac{2m \omega_y dy}{\sqrt{2m \omega  I_y - m^2 \omega_y^2 y^2}} =  \int \frac{dx}{(2I_y /mw_y) - y{^2}}   $$ 

Se realizan las integrales anteriores usando sustitución trigonométrica para obtener $x$ y $y$ en términos de las variables de ángulo. Las sustituciones correspondientes son:

$$ x =  \sqrt{\left ( \frac{2 I_x}{m \omega_x} \right )} sin \theta ; \, \, \, \, \, \, \,  y =  \sqrt{\left ( \frac{2 I_y}{m \omega_y} \right )} sin \theta$$


Dando como resultado:

$$ x  = \sqrt{\left ( \frac{2 I_y}{m \omega_y} \right )} sin \phi_x $$ 

 $$ y  = \sqrt{\left ( \frac{2 I_y}{m \omega_y} \right )} sin \phi_y $$

 Que corresponde a la transformación canónica de las variables.

Ahora, las ecuaciones de Hamilton para las variables de acción - ángulo pueden obtenerse a partir de la acción (la cual corresponde la función generatriz de la transformación) usando (50.5) se tiene que:


$$ \dot{I_x} = - \frac{\partial E}{\partial \phi_x} = 0 ; \, \, \, \dot{I_y} = - \frac{\partial E}{\partial \phi_y} = 0 $$

$$ \dot{\phi_x} = \frac{\partial E}{\partial I_x} = \omega_x $$

y 

$$ \dot{\phi_y} = \frac{\partial E}{\partial I_y} = \omega_y $$

Cuya solución es trivial: 

$$ \phi_x (t) = \phi _{x0} + w_x t $$

$$\phi_y (t) = \phi _{y0} + w_y t$$

Y finalamente, utilziando la transformación canónica se tiene que: 


$$ x  = \sqrt{\left ( \frac{2 I_x}{m \omega_x} \right )} sin (\phi _{x0} + w_x t) $$ 

 $$ y  = \sqrt{\left ( \frac{2 I_y}{m \omega_y} \right )} sin (\phi _{y0} + w_y t) $$


Estas ecuaciones parámetricas corresponden a las figuras de Lissajous, las cuales dependen de la relación entre las frecuencias $\omega_x$ y $\omega_y$, donde se tienen trayectorias periódicas y cerradas si este cociente es un número racional. Adicionalmente se define el área que limita las curvas a partir de el coeficiente que acompaña las soluciones anteriores; estos dependen de las variables de acción (la cual depende de la constante de separación), la masa y la frecuencia de las oscilaciones.

 A continuación se presentan las curvas de Lissajous para distintas relaciones entre las frecuencias: 

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
t = np.linspace(-np.pi, np.pi, 300)

i = 1
m = 0.5
phix = np.pi/2
phiy = np.pi/4*0


def fun(wx, wy, t):
  a = np.sqrt(2*i /(m*wx))
  b =  np.sqrt(2*i /(m*wy))

  x = a*np.sin(wx*t + phix )
  y = b*np.sin(wy*t + phiy)


  s = (x, y)

  return s
```

```python
plt.rcParams['figure.figsize'] = 20, 10 
plt.subplot(2,2,1) # fila,columnas
plt.title('$w_x = 1, w_y = 1$')
plt.plot(fun(1, 1, t)[0], fun(1,1, t)[1], color = 'darkblue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.subplot(2,2,2) 
plt.title('$w_x = 1, w_y = 3$')
plt.plot(fun(1, 3, t)[0], fun(1,3, t)[1], color = 'darkblue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.subplot(2,2,3) 
plt.title('$w_x = 3, w_y = 5$')
plt.plot(fun(3, 5, t)[0], fun(3,5, t)[1], color = 'darkblue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.subplot(2,2,4) 
plt.title('$w_x = 5, w_y = 7$')
plt.plot(fun(5, 7, t)[0], fun(5,7, t)[1], color = 'darkblue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

Veamos que cuando el cociente $w_x / w_y$ no es un número racional, la curva no se cierra. Consideremos $w_x  = \pi$ y $ w_y = 1$

```python
plt.figure(figsize = (8, 6))
plt.title('$w_x = \pi, w_y = e$')
plt.plot(fun(np.pi, np.e, t)[0], fun(np.pi, np.e, t)[1], color = 'darkblue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
