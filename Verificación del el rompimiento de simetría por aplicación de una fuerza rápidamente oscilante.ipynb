Consideremos el siguiente sistema: 


sistema.jpeg

El cual consiste de dos resortes los cuales en su posición natural tiene una longitud $l_0$ y que se deforman en una longitud $d$  (bien sea una contracción o una elongación) hasta una longitud $l$ ($l_0 = l + d$) de tal forma que se unen a los extremos de una pared con una masa $m$ entre ellos.  


Al desplazar la masa una distancia $y$ fuera del eje horizontal, aparece una fuerza de restitución (fuerza lenta) asociada a los dos resortes en dirección vertical. La componente horizontal de la fuerza es anula.  También se considera una fuerza periódica $f(t) = y_0 cos (\gamma t)$ que oscila rápidamente, es decir, su frecuencia es mucho mayor que la frecuencia natural del sistema. Esta fuerza podría lograrse, por ejemplo, con un imán oscilando rápidamente en el eje vertical.

La fuerza neta lenta es:

\begin{equation}
 F = - 2k(s - l_0) sin \theta = -2k(s - l - d)sin\theta = -2sk (sinθ)  + sk(l + d)sin θ \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \,  \, \, \, \, \ (1)
\end{equation}


De la figura se tiene que 

$$sin θ =y/s = \frac{y}{\sqrt{l^2 + y^2}} = \frac{y}{l}  \frac{1}{\sqrt{1 + l^2/y^2}} $$

Expandiendo el último término de la expresión anterior en series de potencia hasta orden 2 se tiene que:

$$ sin θ = \frac{y}{l} \left ( 1 - \frac{y^2}{2l_0^2} \right ) $$ 

Utilizando esto  y operando en la ecuación (1) se obtiene  que:

$$ F = \frac{2kd}{l}y -  \frac{k(l + d)}{l^3}y^3  $$

Y como $l_0 = l + d$: 

$$ F = \frac{2k(l_0 - l)}{l}y -  \frac{kl_0}{l^3}y^3  $$


Integrando la ecuación (2) se obtiene la energía potencial: 

$$ U(y) = - \int F(y) dy  =- \frac{k(l_0 - l)}{l} y^2 +  \frac{kl_0}{4l^3} y^4 $$

$$ U(y)  = - C y^2 +  By^4 \, \, \, \,\, \, \, \,  \, \, \, \,\, \, \, \, \, \, \, \,\, \, \, \, (2) $$ 

Donde $C = \frac{k(l_0 - l)}{l}$ y $B = \frac{kl_0}{4l^3} $ 

A partir de la ecuación (2) correspondiente al potencial, se pueden encontar los puntos de estabilidad. 

$${U(y)}'  = - 2Cy+4By^3 = 0$$
 cuyas raíces corresponden a 

$$y_1 = 0$$
$$y_2 = \pm\sqrt{\frac{C}{2B}} $$

Evaluando ahora la segunda derivada y reemplazando las raíces $y_1$ y $y_2$ se obtiene

$${U''(y_1)}  = - 2C < 0$$

$${U''(y_2)}  = - 2C+12By^2 = -2C +6C >0$$

En este orden de ideas, se consigue un punto de equilibrio estable en $y_2$ y uno inestable en $y_1$.

Considerando ahora que el sistema se somete a una fuerza periódica de frecuancia alta $\gamma$ y de la forma

$$ f(t) = f_0 cos(\gamma t)$$

la ecuación de movimiento del sistema se puede escribir como 

$$ m\ddot{y} = - \frac{dU (y)}{dy} + f(t)$$ (3)

a la cual se suspone una solución de la forma 

$$ y = x + y_0 cos(\gamma t)$$

Donde el primer término de la igualdad está asociado a las osclaciones lentas y el segundo a las rápidas ejercidas por la fuerza externa. En este orden de ideas, tomando la segunda derivada de esta última expresión y sustituyendo en (3) se consigue:

$$m (\ddot{x} - y_0 \gamma ^{2}cos(\gamma t)) = 2C(x + y_0cos(\gamma t)) - 4B(x + y_0 \gamma ^{2}cos(\gamma t))^3$$

tomando ahora el promedio temporal en un periodo a ambos lados de esta igualdad se sigue

$$\left< m(\ddot{x} - y_0 \gamma ^{2}cos(\gamma t)) \right> = \left< 2C(x + y_0cos(\gamma t)) - 4B(x + y_0 \gamma ^{2}cos(\gamma t))^3\right>  $$


$$
\left< m\ddot{x} \right> - \left<  m y_0 \gamma ^{2}cos(\gamma t))\right> = 
\left< 2C(x + y_0cos(\gamma t))\right> - \left< 4B(x + y_0 \gamma ^{2}cos(\gamma t))^3 \right> $$

$$m \ddot{x} = 2Cx - 4Bx^3 - 6Bx y_0^2$$ (4)

donde se ha usado el hecho de que 
 $$ \left< cos(\gamma t) \right> = \left< cos(\gamma t)^3\right> = 0$$
 $$\left< cos(\gamma t)^2\right> = \frac{1}{2}$$

 de la ecuación (4) se encuentra entonces la expresión de la fuerza efectiva

  $$F_{ef} = 2Cx - 4Bx^3 - 6Bx y_0^2 $$

  Integrando esta última expresión encontramos el potencial efectivo

  $$ U_{ef} = - (C -3By_0^{2})x^2 + Bx^4$$
  $$ = -3B (\frac{C}{3B}-T )x^2 + Bx^4$$

  De esta última expresión se pueden realizar las siguientes asignaciones

  $$T = y_0^2$$
  $$T_{c} = \frac{C}{3B}$$

En este orden de ideas

$$ U_{ef} = - 3B(T_{c}-T)x^2 + Bx^4$$



Consideremos ahora el efecto de la fuerza rápidamente oscilante: $f(t) = f_0 cos(\gamma t)$   la cual induce oscilaciones forzadas, donde $f_0 = m(\lambda^2 + γ^2)y0 $, siendo $\lambda = (2C/m)^{1/2} $ .

La ecuación de movimiento es entonces: 

$$ m \ddot{y} = -\frac{dU(y)}{dy} + f(t) $$

# Implementación Númerica

### caso $l = l_0$

Consideremos primero el caso en que $l_0 = l$, es decir, la longitud natural del oscilador coincide con la longitud inicial del sistema. 

Con el potencial obtendio en la ecuación (2),  la ecuación de movimiento es: 

$$ m \ddot{y} = - \frac{k}{4l^2} y^3 + y_0 m γ^2 cos(\gamma t) $$

Cuya solución númerica se implementa escribiendo la ecuación diferencial  ( no lineal )de segundo orden como un sistema de ecuaciones diferenciales de primer orden. Haciendo $u = \frac{dy}{dt}$ y reemplazando se obtiene el siguiente sistema:

$$ u = \frac{dy}{dt} $$
$$ \dot{u} =- \frac{1}{m} \left ( -\frac{k}{4l^2} y^3 + y_0 m \gamma^2 cos(\gamma t)  \right ) $$

Se soluciona dando la condiciones iniciales y utilizando la liberia odeint. 




```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
# constantes 
m = 2                    # masa 
y0 = 5                   # amplitud de la fuerza rápidamente oscilante
k = 85                   # constante elástica 
gamma = 100       # frecuencia de la fuerza oscilante (notar que esta es mucho mayor que la frecuencia natural del sistema)
l = 0.3                  # longitud inicial (coincide con la longitud natural)
B = k/(4*l**2)



# condiciones iniciales de posición y velocidad
y_ini = np.array([0.30, 0])
```

```python
# tiempo de integración
t = np.linspace(0, 5, 10000)

# sistema de ecuaciones diferenciales 
def eq(y, t):
    f1 = y[1]
    f2 = 1/m * ( (-k*(y[0])**3) / (4*l**2) + y0*m*(gamma)*np.cos(gamma*t))
    return [f1, f2]


# solución numérica
y1 = odeint(eq, y_ini, t)
```

```python
plt.figure(figsize = (10, 8))
plt.title('Oscilaciones rápidas del sistema en el caso $l = l_0$', fontsize=15)
plt.plot(t, y1[:, 0], color = 'darkblue')
plt.xlabel('t,', fontsize=15)
plt.ylabel('y(t)', fontsize=15)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
plt.title('Espacio de Fase', fontsize=15)
plt.plot(y1[:, 0], y1[:, 1], color = 'darkblue')
plt.xlabel('y(t)', fontsize=15)
plt.ylabel('$\dot{y(t)}$', fontsize=15)
plt.show()
```

Se oberva como era de esperarse, una oscilación lenta correspondiente a las oscilaciones de la masa conectada con los resortes alrededor de la posición de equilibrio, $y_1 = 0$ y $y_2 = 0$ (pues $C$ = 0 en este caso) más unas oscilaciones rápidas producidas por la fuerza rápidamente oscilante.

# Potencial Efectivo

Del análisis previo se obtuvo que el potencial efectivo se escribe como 

$$ U_{ef} = -3B(T_c - T)x^2 + Bx^4 $$

Donde $T_c = \frac{C}{3B} = 0$ en el caso que estamos considerando ($l = l_0$).  



```python
# Gráfica del potencial Efectivo

x = np.linspace(-0.5, 0.5, 100)
U = lambda x: 3*B*y0**2*x**2 + B*x**4
#U1 = lambda x: 

plt.figure(figsize = (10, 8))
plt.plot(x, U(x), color = 'darkblue')
plt.yticks([])
plt.show()
```

En este caso el potencial es un polinomio de grado 4 que presenta una simetría respecto al eje y.

## Caso $ l \neq l_0 $

En este caos la longitud inicial del sistema $l$, no coincide con la longitud natural de los resortes $l_0$.


La ecuación de movimiento en este caso se escribe como: 

$$m \ddot{y} = 2Cy - 4By^3 + m(λ^2 + γ^2)y_0 cos γ t $$

Donde $B = \frac{k l_0}{4l^3} $; $\, \, $  $C = \frac{k_0 - l}{l}k$ y $\lambda = (2C/m)^{1/2} $

Implementamos la solución númerica similarmente al caso anterior.


Es interesante en este caso, comparar esta solución con la solución obtenida a partir de la ecuación de movimiento en términos de la fuerza efectiva:

$$ m \ddot{x} = 2(C - 3By_0^2)x - 4Bx^3 $$

Donde $x$ es la componente de bajas frecuencias del desplazamiento. 

Resolviendo la ecuación anterior y recordando que $ y = y_0 + cos(\gamma t) $, se obtiene una solución 'efectiva' que se puede comparar con la solución 'normal'.  


```python
# Definición de nuevas constantes
m1 = 2
l0_1 = 1.5
l1 = 0.5
k1 = 5
y0_1 = 0.1
gamma1 = 200


# coeficientes 
B1 = k1*l0_1 / (4*l1**3)
C = k1* (l0_1 - l1)/l1
λ =  (2*C/m1)**(1/2)

Tc = C/(3*B1)
T = y0_1**2

# condiciones iniciales de posición y velocidad
y_ini = np.array([0.30, 0])
x_ini = np.array([0.30, 0])
```

```python
# tiempo de integración
t1 = np.linspace(0, 5, 10000)

# Ecuación 'Normal' 
def eq2(y, t):
  f1 = y[1]
  f2 = 1/m1 * (2*C*y[0] - 4*B1*y[0]**3 + y0_1*m1*(gamma1**2 + λ**2 )*np.cos(gamma1*t))
  return [f1, f2]

# Ecuación 'Efectiva'
def eq3(x, t):
  f1 = x[1]
  f2 = 1/m1 * (2*x[0] * (C - 3*B1*y0_1**2) - 4*B1*x[0]**3)
  return [f1, f2]

# Solución númerica 
sol2 = odeint(eq2, y_ini,  t1)
sol3 = odeint(eq3, x_ini, t1)
```

```python
# Definir una función para calcular y = x + y0 cos γt
def f4(t):
  return sol3[:, 0] + y0_1*np.cos(gamma1*t)
```

```python
# Graficación 
plt.figure(figsize=(10, 8))
plt.title('Oscilaciones rápidas del sistema en el caso $l$ diferente de  $l_0$', fontsize=15)
plt.plot(t1, sol2[:, 0], color = 'darkblue', label = 'Solución Normal')
plt.plot(t1, f4(t1), 'red',  label = 'Solución Efectiva')
plt.xlabel('t', fontsize=15)
plt.ylabel('y(t)', fontsize=15)
plt.legend(fontsize=15)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
plt.title('Espacio de Fase', fontsize=15)
plt.plot(sol2[:, 0], sol2[:, 1], color = 'darkblue', linewidth = 0.5)
plt.xlabel('y(t)', fontsize=15)
plt.ylabel('$\dot{y(t)}$', fontsize=15)
plt.show()
```

Se aprecia, al igual que en el caso anterior, unas oscilaciones lentas asociadas a la oscilación debido a los resortes y unas oscilaciones rápidas debido a la fuerza rapidamente oscilante.

También se observa que la solución normal y la efectiva tienen la misma forma aunque su amplitud de las oscilaciones lentas y rápidas en el tiempo son apreciablemente diferentes, esto se puede deber a las fluctuaciones asociadas a la solución númerica. 

Ahora, el potencial efectivo se escribe como 



$$ U_{ef} = - 3B(T_c - T)x^2 + Bx^4 $$ 

Donde $B = \frac{k l_0}{4l^3} $; $\, \, $  $C = \frac{k_0 - l}{l}k; $ $\, \,$    $T = y _0^2 $ y $ T_c = \frac{C}{3B} $ 

El sistema que estamos estudiando puede llegar a presentar una bifurcación que da lugar a dos puntos de equilibrio estables y uno inestable (cómo se encontró anteriormente) esto, dependiendo de como sea el signo de $T_c - T$. 

Caso $T_c > T$ 

```python
x = np.linspace(-1, 1, 100)
U1 = lambda x: -3*B1*(Tc - T)*x**2 + B1*x**4
```

```python
plt.figure(figsize = (10, 8))
plt.title('Potencial Efectivo para $T_c > T$', fontsize = 15)
plt.plot(x, U1(x), color = 'darkblue')
plt.xlabel('x', fontsize = 15)
plt.ylabel('$U_{ef}$', fontsize = 15)
plt.vlines(0.57, -5, 5, linestyles = 'dashed')
plt.vlines(-0.57, -5, 5, linestyles = 'dashed')
plt.show()
```

En este caso se aprecia precisamente la bifurcación del potencial efectivo en dos pozos de potencial con posiciones estables localizadas en $ x_{\pm } = \pm \left ( \frac{C}{2B} \right ) = \pm 0,57 $ y una posición de equilibrio inestable en $x = 0$. Se dice que en este caso hay un rompimiento de símetria. 

La bifurcación del potencial tiene lugar cuando el origen se destabiliza, es decir, cuando se varía el parámetro $d$; pero de la ecuación del potencial efectivo se tiene que otra forma de realizar lo mismo es variar $T = y_0^2$, es decir, la amplitud de la fuerza de alta ferecuencia.

Variemos este parámetro para unos cuantos casos para observar esta variación: 



```python
# Variación de la amplitud de la fuerza rápidamente oscilante 

xvar = np.linspace(-3, 3, 100)

def Uvar(x, y0):
  return 3*B1*(Tc - y0**2)*x**2 + B1*x**4

y0array = np.array([ 0.5, 0.7, 1, 1.2, 1.5])
```

```python
plt.figure(figsize = (10, 8))
plt.title('Gráfica del Potencial Efectivo para Diferentes valores de T', fontsize = 15)
plt.plot(xvar, Uvar(xvar, y0array[0]),   label = 'T= 0.24')
plt.plot(xvar, Uvar(xvar, y0array[1]),  label = 'T = 0.49')
plt.plot(xvar, Uvar(xvar, y0array[2]),  label = 'T = 1')
plt.plot(xvar, Uvar(xvar, y0array[3]),  label = 'T = 1.44')
plt.plot(xvar, Uvar(xvar, y0array[4]),  label = 'T = 2.25')
plt.xlabel('x', fontsize = 15)
plt.ylabel('$U_{ef}$', fontsize = 15)
plt.legend(fontsize=15)
plt.show()
```
